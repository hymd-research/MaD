```python
from collections import defaultdict
import dpkt
import io
import os
import re
import socket
import zipfile
```

```python
HOME = os.path.expanduser('~')
FILE = f"{HOME}/work/Raw Data/dataleak.pcap"
```

```python
flowList = defaultdict(int)
dataList = []
for count, (ts, buf) in enumerate(dpkt.pcap.Reader(open(FILE, 'rb'))):
    try:
        eth = dpkt.ethernet.Ethernet(buf)
    except:
        print(f"Fail parse. FileNo:{count}, skipped.")
    
    dataList.append(eth)
    if type(eth.data) == dpkt.ip.IP:
        ip = eth.data
        src = socket.inet_ntoa(ip.src)
        dst = socket.inet_ntoa(ip.dst)
        flowWord = src + " to " + dst
        print(flowWord)
        flowList[flowWord] += len(str(buf))
else:
    print(f"{count} files found.")

for key, value in flowList.items():
    print(key, value)
```

```python
request = dataList[0].data.data.data.decode('SHIFT-JIS')
print(request)
```

```python
zipstr = b''
for i in range(1, 9):
    zipstr += dataList[i].data.data.data
zipF = re.search(b'(?<=PK).*(?=------)', zipstr, flags=re.DOTALL)[0]
zipF = bytearray(zipF)
```

```python
zf = zipfile.ZipFile(io.BytesIO(zipF), 'r')
for fileinfo in zf.infolist():
    print(zf.read(fileinfo).decode('ascii'))
```
